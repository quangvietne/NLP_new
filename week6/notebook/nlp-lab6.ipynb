```python
from transformers import pipeline

# 1. Tải pipeline "fill-mask" (thường dùng DistilBERT hoặc BERT)
mask_filler = pipeline("fill-mask", model="bert-base-uncased")

# 2. Câu đầu vào với token [MASK]
input_sentence = "Hanoi is the [MASK] of Vietnam."

# 3. Thực hiện dự đoán (top_k=5 trả về 5 kết quả tốt nhất)
predictions = mask_filler(input_sentence, top_k=5)

# 4. In kết quả
print(f"Câu gốc: {input_sentence}")
for pred in predictions:
    print(f"Dự đoán: '{pred['token_str']}' : {pred['score']:.4f}")
    print(f" -> Câu hoàn chỉnh: {pred['sequence']}")
```

- Câu 1 : Mô hình đã dự đoán đúng từ Capital ( 99%) 
- Câu 2 : mô hình Encoder-only (như BERT) phù hợp  cho tác vụ này vì BERT có kiến trúc hai chiều (bidirectional). Nó có thể nhìn thấy ngữ cảnh từ cả phía trước ("Hanoi is the") và phía sau ("of Vietnam") để suy luận ra từ ở giữa

```python
from transformers import pipeline
# 1. Tải pipeline "text-generation"
# Pipeline này sẽ tự động tải một mô hình phù hợp (thường là GPT-2)
generator = pipeline("text-generation")
# 2. Đoạn văn bản mồi
prompt = "The best thing about learning NLP is"
# 3. Sinh văn bản
# max_length: tổng độ dài của câu mồi và phần được sinh ra
# num_return_sequences: số lượng chuỗi kết quả muốn nhận
generated_texts = generator(prompt, max_length=50, num_return_sequences=1)
# 4. In kết quả
print(f"Câu mồi: '{prompt}'")
for text in generated_texts:
    print("Văn bản được sinh ra:")
    print(text['generated_text'])
```

- Câu 1 : Kết quả sinh ra tương đối hợp lí
- câu 2 : mô hình Decoder-only như GPT lại phù hợp cho tác vụ này vì nó được huấn luyện để dự đoán từ tiếp theo trong một chuỗi. Chúng chỉ có khả năng nhìn “một chiều” (unidirectional), tức là chỉ xem xét các từ đã xuất hiện trước đó.

```python
import torch
from transformers import AutoTokenizer, AutoModel

# 1. Chọn mô hình BERT
model_name = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)

# 2. Câu đầu vào
sentences = ["This is a sample sentence."]

# 3. Tokenize câu
# return_tensors='pt' trả về PyTorch tensor
inputs = tokenizer(sentences, padding=True, truncation=True, return_tensors='pt')

# 4. Đưa qua mô hình để lấy hidden states
with torch.no_grad():
    outputs = model(**inputs)
    # last_hidden_state chứa vector của tất cả các token
    last_hidden_state = outputs.last_hidden_state

# 5. Thực hiện Mean Pooling thủ công
# Lấy attention mask để biết token nào là thật, token nào là padding
attention_mask = inputs['attention_mask']

# Mở rộng chiều của mask để khớp với last_hidden_state
mask_expanded = attention_mask.unsqueeze(-1).expand(last_hidden_state.size()).float()

# Tính tổng các vector (chỉ tính các token thật nhờ nhân với mask)
sum_embeddings = torch.sum(last_hidden_state * mask_expanded, 1)

# Tính tổng số lượng token thật (tránh chia cho 0 bằng clamp)
sum_mask = torch.clamp(mask_expanded.sum(1), min=1e-9)

# Tính trung bình
sentence_embedding = sum_embeddings / sum_mask

# 6. In kết quả
print("Vector biểu diễn của câu:")
print(sentence_embedding)
print("\nKích thước của vector:", sentence_embedding.shape)
```

- Câu 1 : Vector biểu diễn câu có kích thước là 768 , Con số này ứng với hidden_size của mô hình BERT
- Câu 2 : cần sử dụng attention_mask để loại bỏ sự ảnh hưởng của các Token đệm (Padding Tokens) khi tính toán trung bình
